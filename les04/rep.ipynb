```python
import numpy as np
import pandas as pd

# Загрузка данных
df = pd.read_csv('titanic.csv', index_col='PassengerId')

# 1. Первые 6 строк
print("HEAD:")
print(df.head(6))

# 2. Описание данных
print("\nDESCRIBE:")
print(df.describe(include='all'))

# 3. Мужчины / женщины
print("\nSEX COUNT:")
print(df['Sex'].value_counts())

# 4. Распределение Pclass
print("\nPCLASS DISTRIBUTION:")
print(df['Pclass'].value_counts())

print("\nPCLASS BY SEX:")
print(df.groupby('Sex')['Pclass'].value_counts())

# Мужчины 2 класса
men_2_class = len(df[(df['Sex']=='male') & (df['Pclass']==2)])
print("\nMen 2 class:", men_2_class)

# 5. Fare статистика
print("\nFARE STATS:")
print("Median:", round(df['Fare'].median(), 2))
print("Std:", round(df['Fare'].std(), 2))

# 6. <30 vs >60
young = df[df['Age'] < 30]
old = df[df['Age'] > 60]

print("\nSURVIVAL <30 vs >60:")
print("Young:", young['Survived'].mean())
print("Old:", old['Survived'].mean())

# 7. Женщины vs мужчины
print("\nSURVIVAL FEMALE vs MALE:")
print("Female:", df[df['Sex']=='female']['Survived'].mean())
print("Male:", df[df['Sex']=='male']['Survived'].mean())

# 8. Самое популярное мужское имя
male_names = df[df['Sex']=='male']['Name']
first_names = male_names.str.split(',').str[1].str.strip().str.split(' ').str[0]
most_common_name = first_names.value_counts().idxmax()

print("\nMOST POPULAR MALE NAME:", most_common_name)

# 9. Средний возраст по полу и классу
print("\nAGE BY CLASS AND SEX:")
print(df.pivot_table(values='Age', index='Pclass', columns='Sex', aggfunc='mean'))

# Проверка утверждений
men_class1_age = df[(df['Sex']=='male') & (df['Pclass']==1)]['Age'].mean()
women_class1_age = df[(df['Sex']=='female') & (df['Pclass']==1)]['Age'].mean()

print("\nCHECK STATEMENTS:")
print("Men 1 class avg age:", men_class1_age)
print("Women 1 class avg age:", women_class1_age)

# 10. Возраст спасенных vs погибших
survived_age = df[df['Survived']==1]['Age'].mean()
died_age = df[df['Survived']==0]['Age'].mean()

print("\nAGE SURVIVED vs DIED:")
print("Survived:", survived_age)
print("Died:", died_age)
print("Died older?", died_age > survived_age)
```
