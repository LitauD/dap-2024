```python
import sys
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

from sklearn.datasets import load_iris

print(f"Python version: {sys.version}")
print(f"Numpy version: {np.version.version}")
print(f"Matplotlib version: {mpl.__version__}")

# =========================
# ЗАДАЧА 1
# =========================
x1 = np.linspace(-10, 0, 100)
x2 = np.linspace(0, 7, 100)

y1 = x1**2
y2 = x2**2

plt.figure()
plt.plot(x1, y1, 'r-.', label='x^2 [-10,0]')
plt.plot(x2, y2, color=(1.0,0.2,0.3), label='x^2 [0,7]')

plt.xlim(-10,10)
plt.ylim(-10,10)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Задача 1")
plt.legend()
plt.show()

# =========================
# ЗАДАЧА 2
# =========================
df = pd.read_csv('titanic.csv')
df = df.dropna(subset=['Age','Fare'])

surv = df[df['Survived']==1]
dead = df[df['Survived']==0]

plt.figure()
plt.scatter(surv['Age'], surv['Fare'], c='green', marker='o', label='Survived')
plt.scatter(dead['Age'], dead['Fare'], c='red', marker='x', label='Dead')

plt.xlabel("Age")
plt.ylabel("Fare")
plt.title("Titanic Scatter")
plt.legend()
plt.show()

# =========================
# ЗАДАЧА 3
# =========================
iris = load_iris()
X = iris.data
y = iris.target
names = iris.feature_names

fig, axes = plt.subplots(4,4, figsize=(12,12))

for i in range(4):
    for j in range(4):
        ax = axes[i,j]
        ax.scatter(X[:,j], X[:,i], c=y)
        ax.set_xlabel(names[j])
        ax.set_ylabel(names[i])

plt.suptitle("Iris Scatter Matrix")
plt.tight_layout()
plt.show()

# =========================
# ЗАДАЧА 4
# =========================
df = pd.read_csv('titanic.csv')

plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
df['Age'].dropna().hist(bins=20)
plt.title("Age Distribution")
plt.xlabel("Age")

plt.subplot(1,2,2)
df[df['Sex']=='female']['Fare'].dropna().hist(bins=20)
plt.title("Fare (Female)")
plt.xlabel("Fare")

plt.tight_layout()
plt.show()

# =========================
# ЗАДАЧА 5
# =========================
df = pd.read_csv('titanic.csv')

mean_fare = df.groupby('Pclass')['Fare'].mean()

plt.figure()
mean_fare.plot(kind='bar')
plt.xlabel("Pclass")
plt.ylabel("Mean Fare")
plt.title("Mean Fare by Class")
plt.show()

# =========================
# ЗАДАЧА 6
# =========================
count = df.groupby(['Pclass','Sex']).size().unstack()

count.plot(kind='bar', stacked=True)
plt.xlabel("Pclass")
plt.ylabel("Count")
plt.title("Passengers by Class and Sex")
plt.show()

# =========================
# ЗАДАЧА 7
# =========================
iris = load_iris()
counts = np.bincount(iris.target)

plt.figure()
explode = [0.1 if i==np.argmax(counts) else 0 for i in range(3)]
plt.pie(counts, labels=iris.target_names, explode=explode, autopct='%1.1f%%')
plt.title("Iris Distribution")
plt.show()

# =========================
# ЗАДАЧА 8
# =========================
df = pd.read_csv('titanic.csv')

sns.histplot(df['Age'].dropna(), bins=20)
plt.title("Age Distribution (Seaborn)")
plt.show()
```
