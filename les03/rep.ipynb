```python
import numpy as np

# 1. Массив 10x10, минимум и максимум
z = np.random.rand(10,10)
print("Min:", z.min(), "Max:", z.max())


# 2. Шахматная матрица 8x8
z = np.zeros((8,8))
z[1::2, ::2] = 1
z[::2, 1::2] = 1
print(z)


# 3. Поменять знак элементов между 3 и 8
z = np.arange(11)
z[(z > 3) & (z < 8)] *= -1
print(z)


# 4. Вычесть среднее строки
z = np.random.rand(5,10)
z = z - z.mean(axis=1, keepdims=True)
print(z)


# 5. Поменять 1 и 2 строки
z = np.arange(25).reshape(5,5)
z[[0,1]] = z[[1,0]]
print(z)


# 6. Декартовы → полярные
z = np.random.random((10,2))
x, y = z[:,0], z[:,1]
r = np.sqrt(x**2 + y**2)
theta = np.arctan2(y, x)
polar = np.column_stack((r, theta))
print(polar)


# 7. Заменить максимум на 0
z = np.random.random(10)
z[z.argmax()] = 0
print(z)


# 8. Добавить 1 по индексам (только 1 раз)
z = np.ones(10)
i = np.random.randint(0,len(z),20)
z += np.bincount(i, minlength=len(z))
print(z)


# 9. Вектор с нулями между элементами
z = np.array([1,2,3,4,5])
result = np.zeros(len(z)*4 - 3)
result[::4] = z
print(result)


# 10. Ближайшее значение
z = np.arange(100)
v = np.random.uniform(0,100)
nearest = z[np.abs(z - v).argmin()]
print("v =", v, "nearest =", nearest)
```
